### 6. Creando la funci√≥n de regresi√≥n lineal 

Los argumentos de a funci√≥n deben ser:

6.1 Vector con la variable independiente __x__

6.2 Vector con la variable dependiente **y**

6.3 un entero **epochs** que indica por cuantas iteraciones entrenar el modelo.

6.4 un entero **imprimir_error_cada** , que nos indica cada cuantas iteraciones queremos imprimir a trav√©s de print: el n√∫mero de iteraci√≥n, el error del modelo en esa iteraci√≥n, si imprimir error cada = 10, se despliega en pantalla el error en las iteraciones:10,20,30,40,50.

6.5 Escalar $\alpha$ (learning rate): es usado como parte de la expresi√≥n matem√°tica para actualizar en cada iteraci√≥n los par√°metros del modelo.
    

```python
def grafico_error(errores):
    plt.scatter(np.array(range(len(errores))), errores,c='red',s=3)
    plt.title('Gr√°fica de error para cada interacci√≥n')
    plt.xlabel('Epoch')
    plt.ylabel('Error')
    plt.show()
```

```python
def regresion(varX,varY,epoch,imprimir_error_cada,lr, m= 0, b=0):
       
    params = np.array([[m],[b]])
    vec_error=[0.0]*epoch # vector de errores
    dict_modelo = {} # inicializaci√≥n de diccionario para almacenar modelo (m,b)
        
    y = np.array(varY) #Conversi√≥n de la variable Y a un vector
    x = np.vstack(varX) #Conversi√≥n de la variable X a un vector vertical
    matriz_x = np.concatenate((x, np.ones_like(x)), axis = 1) #Creando la matriz de n*2
   
    for i in range(epoch):
        y_hat = np.dot(matriz_x,params) # producto punto calcular el valor estimado, y = mX + b
        error = 0.5 * np.mean((y - y_hat)**2) # c√°lculo del error
        vec_error[i] = error # almacenamiento del error en un vector
        
        vec_gradiente = np.mean(matriz_x*(y_hat - y),axis=0) # vector_gradiente (gradiente_m, gradiente_b)
        
        # C√°lculo de par√°metros del modelo
        m = params[0] - lr * vec_gradiente[0]       # actualizando m
        b = params[1] - lr * vec_gradiente[1]       # actualizando b
        
        # Almacenamiendo los par√°metros (m,b)
        dict_modelo[i] = [np.around(m[0],2),np.around(b[0],2)]
        
        # Actualizaci√≥n de par√°metros del modelo para la siguiente interaci√≥n
        params[0] = m
        params[1] = b
        
        # Visualizaci√≥n de progresi√≥n de errores y sus respectivos par√°metros m y b
        if (i+1) % imprimir_error_cada == 0 or i == 0:
            print("EPOCH ",i+1,": Error =",np.around(error,2)," [m,b]=",dict_modelo[i])
    
    
    # Llamando la funci√≥n para graficar el error
    grafico_error(vec_error)
    
    print("Estimaci√≥n Final de par√°metros:")
    print("m: ",np.around(params[0,0],2))
    print("b: ",np.around(params[1,0],2))
    print("p√©rdida m√≠nima :",np.around(error,2))
    
    def funcion(x):
        return params[0,0] * x + params[1,0]

    return dict_modelo,vec_error,funcion
```

```python
matrixtrain = np.array(train)
```

```python
SalePrice = matrixtrain[:,[0]];
OverallQual = matrixtrain[:,[1]];
FstFlrSF = matrixtrain[:,[2]];
```

Usando el modelo de entrenamiento para __OverallQual__ como como variable predictora 

```python
dict_OverallQual = {}
error_OverallQual = []

dict_OverallQual,error_OverallQual,regresion_OverallQual= regresion(OverallQual,SalePrice,1000,100,0.00001)
```

### Evoluci√≥n del modelo entrenado en el tiempo

Se crea una funci√≥n para observar la evoluci√≥n del modelo en el tiempo con cada "n" iteraciones y las grafica en conjunto con los datos de entrenamiento.

##### Par√°metros de entrada:
* Diccionario de valores con los par√°metros [m,b] del modelo
* Vector de Valores "X": variable predictora
* Vector de Valores "Y": variable a predecir
* Frecuencia de Iteraci√≥n a Mostrar

```python
def evolucion(diccionario,x,y,n):
    
    modelo_i = np.empty_like(x)
    
    print("Entrenamiento:")
        
    for i in range(len(diccionario)): # Para cada iteraci√≥n en el Diccionario de Par√°metros m,b
        
        if (i+1) % n == 0 or i == 0: # Cada "n" iteraciones...
            
            for j in range(len(x)): # Se calcula el modelo: y_hat's correspondientes para todas las "x"
                modelo_i[j] = diccionario[i][0]*x[j] + diccionario[i][1]   # y_hat = m*x + b                
            
            # Se muestran detalles del modelo y se dibuja la recta con Matplotlib
            print("Iteraci√≥n",i+1,": y = ",diccionario[i][0],"*x + ",diccionario[i][1])           
            plt.plot(x,modelo_i,label='Iteraci√≥n No. {i}'.format(i=i+1))
            plt.legend(loc = 'best')
                
    print("")
    plt.scatter(x,y,c='olive',s=10)
    plt.title('Evoluci√≥n del Entrenamiento')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    return
```
